`__add__()` lets us teach Python what the `+` operator means for our custom objects, making our code cleaner, more intuitive, and more powerful.

### Scenario  
Money class that only adds amount (currency must match). We only care about the amount, not the currency symbol.

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency

m1 = Money(50, '$')
m2 = Money(35, '$')

mtotal = m1 + m2
```

Sure an error will raise since python does not know what to add. Here is the fix:

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency
    
    def __add__(self, other):
        return self.amount + other.amount
    

m1 = Money(50, '$')
m2 = Money(45, '$')
 
print(m1+m2)
```
